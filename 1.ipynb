```python
import numpy as np
import pandas as pd
data=pd.read_csv("enjoysport.csv")
data.head()
s=[""]*4
h=[["?" for i in range (6)] for i in range(6)]
a=[]
for i in range(len(data)):
    a.append(data.iloc[i].tolist())
s=a[0]
for i in range(4):
    if a[i][6]=="Yes":
        for j in range(6):
            if s[j]!=a[i][j]:
                s[j]="?"
                h[j][j]="?"
    else:
        for k in range(6):
            if s[k]!=a[i][k]:
                h[k][k]=s[k]
            else:
                h[k][k]="?"
count=h.count(['?','?','?','?','?','?'])
for i in range(count):
    h.remove(['?','?','?','?','?','?'])
print(s)
print(h)
```
